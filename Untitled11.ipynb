```python
import pandas as pd
a=pd.DataFrame([("Abhi","Lam",20),
                ("vamsi","Ga",22),
                ("giri","galal",30),
                ("aki","kala")],
               columns=("Name","Initial","Age"))
a
```

```python
import pandas as pd
b=pd.DataFrame([("Abhi","Lam",20),
                ("ajsy","Guntu",22),
                ("srinu","guddu",30)],
               columns=("Name","Initial","Age"))
b
```

```python
a.append(b)
```

```python
a.append(b,ignore_index=True,verify_integrity=True)
```

```python
a.append(b,ignore_index=True,sort=True,verify_integrity=False)
```

MERGE

```python
df=pd.merge(a,b,on="Age")
df
```

```python
df=pd.merge(a,b,on="Name")
df
```

```python
df=pd.merge(a,b,on="Initial")
df
```

```python
df=pd.merge(a,b,how="inner")
df
```

```python
df=pd.merge(a,b,how="outer",indicator=True)
df
```

```python
df=pd.merge(a,b,how="left",indicator=True)
df
```

```python
df=pd.merge(a,b,how="right")
df
```

```python
df=pd.merge(a,b,on="Age",suffixes=('_left',"_right"))
df
```

CONCATENATION

```python
pd.concat([a,b],ignore_index=True)
```

```python
pd.concat([a,b],keys=["india","ak"])
```

```python
df=pd.concat([a,b],axis=1)
df
```

```python
pd.concat([a,b],ignore_index=True,join="inner",axis=1)
```

```python
pd.concat([a,b],ignore_index=True,join="inner",axis=0)
```

```python
pd.concat([a,b],ignore_index=True,join="outer",axis=1)
```
